```python
# Imports
import matplotlib.pyplot as plt
import numpy as np
'''
<!-- B: La Suma de Riemann del punto medio. -->
<!-- La funci贸n a la cual se le debe aproximar el 谩rea bajo la curva es
() = ^(/2)+1
En el intervalo [0,3]. Y las aproximaciones que se deben generar est谩n
determinadas por las particiones de la siguiente tabla con su
respectivo error porcentual. -->
### B: La Suma de Riemann del punto medio.
La funci贸n a la cual se le debe aproximar el 谩rea bajo la curva es $f(x) = e^{x/2}+1$ en el intervalo $[0,3]$. Y las aproximaciones que se deben generar est谩n determinadas por las particiones de la siguiente tabla con su respectivo error porcentual.
<!-- Tabla de aproximaciones
Partici贸n(n) | Suma Riemann | Error(%)
4            |              |
10           |              |
800          |              |
5000         |              |
 -->
| Partici贸n(n) | Suma Riemann | Error(%) |
|--------------|--------------|----------|
| 4            |              |          |
| 10           |              |          |
| 800          |              |          |
| 5000         |              |          |

<!-- Posteriormente, adjunte cuatro gr谩ficas realizadas en python de las
4 aproximaciones realizadas en la tabla con su respectivo error
relativo porcentual. -->

Posteriormente, adjunte cuatro gr谩ficas realizadas en python de las 4 aproximaciones realizadas en la tabla con su respectivo error relativo porcentual.

'''

euler = np.e


# Funciones
def f(x):
    return euler**(x/2)+1

# integral de f(x) en [a, b]
def integral(f, a, b):
    return np.trapz([f(x) for x in np.linspace(a, b, 1000)], np.linspace(a, b, 1000))

def riemann(f, a, b, n):
    h = (b-a)/n
    suma = 0
    for i in range(n):
        suma += f(a+i*h)
    return suma*h

def error(f, a, b, n):
    return abs((integral(f, a, b)-riemann(f, a, b, n))/integral(f, a, b))*100

def grafica(f, a, b, n):
    # Grafica con n particiones, indicar la mitad de cada cuadro, colocar linea de funcion
    h = (b-a)/n # ancho de cada cuadro
    x = [a+i*h for i in range(n+1)] # puntos de inicio de cada cuadro
    y = [f(x[i]+h/2) for i in range(n+1)] # puntos de inicio de cada cuadro
    plt.plot([a, b], [0, 0], 'k')
    for i in range(n):
        plt.plot([x[i], x[i]], [0, y[i]], 'b')
        plt.plot([x[i], x[i+1]], [y[i], y[i]], 'b')
        plt.plot([x[i+1], x[i+1]], [0, y[i]], 'b')
        # mitad de cada cuadro
        plt.plot([x[i]+h/2, x[i]+h/2], [0, y[i]], 'r')
    # linea de la funcion
    plt.plot(np.linspace(a, b, 1000), [f(x) for x in np.linspace(a, b, 1000)], 'g')
    plt.show()

# Particiones
n = [4, 10, 800, 5000]
for i in n:
    print('Partici贸n: {} | Suma Riemann: {} | Error: {}'.format(i, riemann(f, 0, 3, i), error(f, 0, 3, i)))
    grafica(f, 0, 3, i)

# Imprimir tabla de resultados
# Encabezados
print('| Partici贸n(n) | Suma Riemann | Error(%) |')
print('|--------------|--------------|----------|')
# Filas
for i in n:
    print('| {} | {} | {} |'.format(i, riemann(f, 0, 3, i), error(f, 0, 3, i)))



```

#### 1. Revisi贸n te贸rica de Series de Taylor

Una serie de Taylor es una serie infinita que representa una funci贸n en un punto dado. La serie de Taylor de una funci贸n $f(x)$ alrededor de un punto $a$ es la siguiente:

$$f(x) = \sum_{n=0}^{\infty} \frac{f^{(n)}(a)}{n!}(x-a)^n$$

donde $f^{(n)}(a)$ es la derivada $n$-茅sima de $f(x)$ evaluada en $a$.

```python
# ### M茅todo de la bisecci贸n


# Se define la funci贸n $xsen(x)$:   
def f(x):
    return x*np.sin(x) - 0.98

# Intervalo inicial:
a = -10
b = 10

# Se define la tolerancia:
tol = 0.000001

# Se define el n煤mero m谩ximo de iteraciones:
N = 100


# Se define la funci贸n que realiza el m茅todo de la bisecci贸n:
def biseccion(f, a, b, tol, N):
    i = 1
    while i <= N:
        p = (a + b)/2
        if f(p) == 0 or (b - a)/2 < tol:
            return p
        i = i + 1
        if f(p)*f(a) > 0:
            a = p
        else:
            b = p
    print("El m茅todo fracas贸 en", N, "iteraciones")


# Se imprime el resultado:
print("La ra铆z es:", biseccion(f, a, b, tol, N))


```

### C贸digo Python para el m茅todo de Newton-Raphson
Se presenta a continuaci贸n el c贸digo Python para el m茅todo de Newton-Raphson:

```python
# ### M茅todo de Newton-Raphson
# Escribir un programa llamado Newton que tenga como argumentos de entrada la funci贸n  f , su derivada  df , el punto inicial  x0 , la cota de error absoluto  tol  y el n煤mero m谩ximo de iteraciones  maxiter  y como argumentos de salida la soluci贸n aproximada y el n煤mero de iteraciones realizadas.

# Utilizarlo para aproximar las tres ra铆ces de la funci贸n f(x)=xsen(x)0.98  en el intervalo  [10,10]  con una precisi贸n de  106  y comparar los resultados con los obtenidos en el apartado anterior.

# Se define la funci贸n $xsen(x)$:
def f(x):
    return x*np.sin(x) - 0.98

# Se define la derivada de la funci贸n $xsen(x)$:
def df(x):
    return np.sin(x) + x*np.cos(x)

# Intervalo inicial:
a = -10
b = 10

# Se define el punto inicial:
x0 = 0

# Se define la tolerancia:
tol = 0.000001

# Se define el n煤mero m谩ximo de iteraciones:
N = 100


# Se define la funci贸n que realiza el m茅todo de Newton-Raphson:
def Newton(f, df, x0, tol, N):
    i = 1
    while i <= N:
        p = x0 - f(x0)/df(x0)
        if abs(p - x0) < tol:
            return p
        i = i + 1
        x0 = p
    print("El m茅todo fracas贸 en", N, "iteraciones")


# Se imprime el resultado:
print("La ra铆z es:", Newton(f, df, x0, tol, N))



```
