```python
# Imports
import matplotlib.pyplot as plt
import numpy as np
'''
<!-- B: La Suma de Riemann del punto medio. -->
<!-- La función a la cual se le debe aproximar el área bajo la curva es
𝑓(𝑥) = 𝑒^(𝑥/2)+1
En el intervalo [0,3]. Y las aproximaciones que se deben generar están
determinadas por las particiones de la siguiente tabla con su
respectivo error porcentual. -->
### B: La Suma de Riemann del punto medio.
La función a la cual se le debe aproximar el área bajo la curva es $f(x) = e^{x/2}+1$ en el intervalo $[0,3]$. Y las aproximaciones que se deben generar están determinadas por las particiones de la siguiente tabla con su respectivo error porcentual.
<!-- Tabla de aproximaciones
Partición(n) | Suma Riemann | Error(%)
4            |              |
10           |              |
800          |              |
5000         |              |
 -->
| Partición(n) | Suma Riemann | Error(%) |
|--------------|--------------|----------|
| 4            |              |          |
| 10           |              |          |
| 800          |              |          |
| 5000         |              |          |

<!-- Posteriormente, adjunte cuatro gráficas realizadas en python de las
4 aproximaciones realizadas en la tabla con su respectivo error
relativo porcentual. -->

Posteriormente, adjunte cuatro gráficas realizadas en python de las 4 aproximaciones realizadas en la tabla con su respectivo error relativo porcentual.

'''

euler = np.e


# Funciones
def f(x):
    return euler**(x/2)+1

# integral de f(x) en [a, b]
def integral(f, a, b):
    return np.trapz([f(x) for x in np.linspace(a, b, 1000)], np.linspace(a, b, 1000))

def riemann(f, a, b, n):
    h = (b-a)/n
    suma = 0
    for i in range(n):
        suma += f(a+i*h)
    return suma*h

def error(f, a, b, n):
    return abs((integral(f, a, b)-riemann(f, a, b, n))/integral(f, a, b))*100

def grafica(f, a, b, n):
    # Grafica con n particiones, indicar la mitad de cada cuadro, colocar linea de funcion
    h = (b-a)/n # ancho de cada cuadro
    x = [a+i*h for i in range(n+1)] # puntos de inicio de cada cuadro
    y = [f(x[i]+h/2) for i in range(n+1)] # puntos de inicio de cada cuadro
    plt.plot([a, b], [0, 0], 'k')
    for i in range(n):
        plt.plot([x[i], x[i]], [0, y[i]], 'b')
        plt.plot([x[i], x[i+1]], [y[i], y[i]], 'b')
        plt.plot([x[i+1], x[i+1]], [0, y[i]], 'b')
        # mitad de cada cuadro
        plt.plot([x[i]+h/2, x[i]+h/2], [0, y[i]], 'r')
    # linea de la funcion
    plt.plot(np.linspace(a, b, 1000), [f(x) for x in np.linspace(a, b, 1000)], 'g')
    plt.show()

# Particiones
n = [4, 10, 800, 5000]
for i in n:
    print('Partición: {} | Suma Riemann: {} | Error: {}'.format(i, riemann(f, 0, 3, i), error(f, 0, 3, i)))
    grafica(f, 0, 3, i)

# Imprimir tabla de resultados
# Encabezados
print('| Partición(n) | Suma Riemann | Error(%) |')
print('|--------------|--------------|----------|')
# Filas
for i in n:
    print('| {} | {} | {} |'.format(i, riemann(f, 0, 3, i), error(f, 0, 3, i)))



```

#### 1. Revisión teórica de Series de Taylor

Una serie de Taylor es una serie infinita que representa una función en un punto dado. La serie de Taylor de una función $f(x)$ alrededor de un punto $a$ es la siguiente:

$$f(x) = \sum_{n=0}^{\infty} \frac{f^{(n)}(a)}{n!}(x-a)^n$$

donde $f^{(n)}(a)$ es la derivada $n$-ésima de $f(x)$ evaluada en $a$.

### Código Python para el método de la bisección
Se presenta a continuación el código Python para el método de la bisección:


```python
# ### Método de la bisección


# Se define la función $xsen(x)$:   
def f(x):
    return x*np.sin(x) - 0.98

# Intervalo inicial:
a = -10
b = 10

# Se define la tolerancia:
tol = 0.000001

# Se define el número máximo de iteraciones:
N = 100


# Se define la función que realiza el método de la bisección:
def biseccion(f, a, b, tol, N):
    i = 1
    while i <= N:
        p = (a + b)/2
        if f(p) == 0 or (b - a)/2 < tol:
            return p
        i = i + 1
        if f(p)*f(a) > 0:
            a = p
        else:
            b = p
    print("El método fracasó en", N, "iteraciones")


# Se imprime el resultado:
print("La raíz es:", biseccion(f, a, b, tol, N))


```

### Código Python para el método de Newton-Raphson
Se presenta a continuación el código Python para el método de Newton-Raphson:

```python
# ### Método de Newton-Raphson
# Escribir un programa llamado Newton que tenga como argumentos de entrada la función  f , su derivada  df , el punto inicial  x0 , la cota de error absoluto  tol  y el número máximo de iteraciones  maxiter  y como argumentos de salida la solución aproximada y el número de iteraciones realizadas.

# Utilizarlo para aproximar las tres raíces de la función f(x)=xsen(x)−0.98  en el intervalo  [−10,10]  con una precisión de  10−6  y comparar los resultados con los obtenidos en el apartado anterior.

# Se define la función $xsen(x)$:
def f(x):
    return x*np.sin(x) - 0.98

# Se define la derivada de la función $xsen(x)$:
def df(x):
    return np.sin(x) + x*np.cos(x)

# Intervalo inicial:
a = -10
b = 10

# Se define el punto inicial:
x0 = 0

# Se define la tolerancia:
tol = 0.000001

# Se define el número máximo de iteraciones:
N = 100


# Se define la función que realiza el método de Newton-Raphson:
def Newton(f, df, x0, tol, N):
    i = 1
    while i <= N:
        p = x0 - f(x0)/df(x0)
        if abs(p - x0) < tol:
            return p
        i = i + 1
        x0 = p
    print("El método fracasó en", N, "iteraciones")


# Se imprime el resultado:
print("La raíz es:", Newton(f, df, x0, tol, N))



```
