#### 1. Revisi√≥n te√≥rica de Series de Taylor

Una serie de Taylor es una serie infinita que representa una funci√≥n en un punto dado. La serie de Taylor de una funci√≥n $f(x)$ alrededor de un punto $a$ es la siguiente:

$$f(x) = \sum_{n=0}^{\infty} \frac{f^{(n)}(a)}{n!}(x-a)^n$$

donde $f^{(n)}(a)$ es la derivada $n$-√©sima de $f(x)$ evaluada en $a$.


#### 2. Desarrollo de un programa Python que determine los polinomios Taylor desde el grado cero (0) hasta el grado cinco (5) usando celdas de c√≥digo y sustentando las l√≠neas de c√≥digo de cada paso con comentarios.

Para este ejercicio se utilizar√° la funci√≥n $f(x) = e^{x^2}$, centrada en $a = 0$ y evaluada en $x = 0.5$.



```python
import matplotlib.pyplot as plt
import numpy as np

'''
<!-- Ejercicio 2: Teor√≠a de Errores -->
<!-- B: ùëì(ùë•) = ùëí^(ùë•^2), centrado en ùëé = 0 y evaluado en ùë• = 0.5 -->
## B: Teor√≠a de Errores

<!-- 
‚Ä¢ Desarrollo de un programa Python que determine los polinomios
Taylor desde el grado cero (0) hasta el grado cinco (5) usando
celdas de c√≥digo y sustentando las l√≠neas de c√≥digo de cada
paso con comentarios.
‚Ä¢ Realice una gr√°fica comparativa con la evaluaci√≥n de los
polinomios solicitados en el √≠tem anterior y determine cu√°l es el
error de truncamiento resultante en cada uno de los polinomios
hallados. Haga una gr√°fica del comportamiento del error para
cada polinomio.-->

'''

# Definici√≥n de la funci√≥n
def f(x):
    return np.exp(x**2)

# Definici√≥n de la derivada n-√©sima de la funci√≥n
def dx(f):
    def df(x):
        return f(x+1e-5)-f(x-1e-5)
    return df
print('Funci√≥n f(x) = e^(x^2):')
print(f(0.5))
print('Derivada de la funci√≥n:')
# grado 0
print('Grado 0:')
print(f(0.5))
# grado 1
print('Grado 1:')
print(dx(f)(0.5))
# grado 2
print('Grado 2:')
print(dx(dx(f))(0.5))
# grado 3
print('Grado 3:')
print(dx(dx(dx(f)))(0.5))
# grado 4
print('Grado 4:')
print(dx(dx(dx(dx(f))))(0.5))
# grado 5
print('Grado 5:')
print(dx(dx(dx(dx(dx(f)))))(0.5))

# Gr√°fica de la funci√≥n
x = np.linspace(-1,1,100)
plt.plot(x,f(x),label='f(x)')
plt.legend()
plt.show()

# Gr√°fica de los polinomios
x = np.linspace(-1,1,100)
plt.plot(x,f(x),label='f(x)')
plt.plot(x,f(0.5)+dx(f)(0.5)*(x-0.5),label='P1(x)')
plt.plot(x,f(0.5)+dx(f)(0.5)*(x-0.5)+dx(dx(f))(0.5)*(x-0.5)**2/2,label='P2(x)')
plt.plot(x,f(0.5)+dx(f)(0.5)*(x-0.5)+dx(dx(f))(0.5)*(x-0.5)**2/2+dx(dx(dx(f)))(0.5)*(x-0.5)**3/6,label='P3(x)')
plt.plot(x,f(0.5)+dx(f)(0.5)*(x-0.5)+dx(dx(f))(0.5)*(x-0.5)**2/2+dx(dx(dx(f)))(0.5)*(x-0.5)**3/6+dx(dx(dx(dx(f))))(0.5)*(x-0.5)**4/24,label='P4(x)')
plt.plot(x,f(0.5)+dx(f)(0.5)*(x-0.5)+dx(dx(f))(0.5)*(x-0.5)**2/2+dx(dx(dx(f)))(0.5)*(x-0.5)**3/6+dx(dx(dx(dx(f))))(0.5)*(x-0.5)**4/24+dx(dx(dx(dx(dx(f)))))(0.5)*(x-0.5)**5/120,label='P5(x)')
plt.legend()
plt.show()

# Gr√°fica del error
x = np.linspace(-1,1,100)
plt.plot(x,f(x)-f(0.5),label='f(x)')
plt.plot(x,f(0.5)+dx(f)(0.5)*(x-0.5)-f(0.5),label='P1(x)')
plt.plot(x,f(0.5)+dx(f)(0.5)*(x-0.5)+dx(dx(f))(0.5)*(x-0.5)**2/2-f(0.5),label='P2(x)')
plt.plot(x,f(0.5)+dx(f)(0.5)*(x-0.5)+dx(dx(f))(0.5)*(x-0.5)**2/2+dx(dx(dx(f)))(0.5)*(x-0.5)**3/6-f(0.5),label='P3(x)')
plt.plot(x,f(0.5)+dx(f)(0.5)*(x-0.5)+dx(dx(f))(0.5)*(x-0.5)**2/2+dx(dx(dx(f)))(0.5)*(x-0.5)**3/6+dx(dx(dx(dx(f))))(0.5)*(x-0.5)**4/24-f(0.5),label='P4(x)')
plt.plot(x,f(0.5)+dx(f)(0.5)*(x-0.5)+dx(dx(f))(0.5)*(x-0.5)**2/2+dx(dx(dx(f)))(0.5)*(x-0.5)**3/6+dx(dx(dx(dx(f))))(0.5)*(x-0.5)**4/24+dx(dx(dx(dx(dx(f)))))(0.5)*(x-0.5)**5/120-f(0.5),label='P5(x)')
plt.legend()
plt.show()


```
