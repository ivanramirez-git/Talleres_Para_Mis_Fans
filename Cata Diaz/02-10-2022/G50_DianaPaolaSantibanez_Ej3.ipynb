### C√≥digo Python para el m√©todo de la bisecci√≥n
Se presenta a continuaci√≥n el c√≥digo Python para el m√©todo de la bisecci√≥n:


```python
# ### M√©todo de la bisecci√≥n


# Se define la funci√≥n $xsen(x)$:   
def f(x):
    return x*np.sin(x) - 0.98

# Intervalo inicial:
a = -10
b = 10

# Se define la tolerancia:
tol = 0.000001

# Se define el n√∫mero m√°ximo de iteraciones:
N = 100


# Se define la funci√≥n que realiza el m√©todo de la bisecci√≥n:
def biseccion(f, a, b, tol, N):
    i = 1
    while i <= N:
        p = (a + b)/2
        if f(p) == 0 or (b - a)/2 < tol:
            return p
        i = i + 1
        if f(p)*f(a) > 0:
            a = p
        else:
            b = p
    print("El m√©todo fracas√≥ en", N, "iteraciones")


# Se imprime el resultado:
print("La ra√≠z es:", biseccion(f, a, b, tol, N))


```

### C√≥digo Python para el m√©todo de Newton-Raphson
Se presenta a continuaci√≥n el c√≥digo Python para el m√©todo de Newton-Raphson:

```python
# ### M√©todo de Newton-Raphson
# Escribir un programa llamado Newton que tenga como argumentos de entrada la funci√≥n  f , su derivada  df , el punto inicial  x0 , la cota de error absoluto  tol  y el n√∫mero m√°ximo de iteraciones  maxiter  y como argumentos de salida la soluci√≥n aproximada y el n√∫mero de iteraciones realizadas.

# Utilizarlo para aproximar las tres ra√≠ces de la funci√≥n f(x)=xsen(x)‚àí0.98  en el intervalo  [‚àí10,10]  con una precisi√≥n de  10‚àí6  y comparar los resultados con los obtenidos en el apartado anterior.

# Se define la funci√≥n $xsen(x)$:
def f(x):
    return x*np.sin(x) - 0.98

# Se define la derivada de la funci√≥n $xsen(x)$:
def df(x):
    return np.sin(x) + x*np.cos(x)

# Intervalo inicial:
a = -10
b = 10

# Se define el punto inicial:
x0 = 0

# Se define la tolerancia:
tol = 0.000001

# Se define el n√∫mero m√°ximo de iteraciones:
N = 100


# Se define la funci√≥n que realiza el m√©todo de Newton-Raphson:
def Newton(f, df, x0, tol, N):
    i = 1
    while i <= N:
        p = x0 - f(x0)/df(x0)
        if abs(p - x0) < tol:
            return p
        i = i + 1
        x0 = p
    print("El m√©todo fracas√≥ en", N, "iteraciones")


# Se imprime el resultado:
print("La ra√≠z es:", Newton(f, df, x0, tol, N))



```
